```python
import numpy as np
import matplotlib as mpl

def j3_s6_intersect(gd_track, roi, slope_threshold=0.01):
    # shapefile of j3 and s6 tracks have horizontal lines, which are are not actually the orbit of these satellites.
    # the multiline feature that denotes the actual ground track of the orbit is split at 180 longitude. The horizontal line is the result of this split, joining the two ends of the split geometry.
    # a simple intersection of the roi with the gd_track will not work because of this. So we need to also check if the roi intersects with this artifant. Checking the slope of the intersect can help us determine if it is the actual orbit or the artifact.
    naive_intersect = gd_track[gd_track.intersects(roi.geometry.values[0])]

    if len(naive_intersect) != 0:
        intersection_geometry = naive_intersect.geometry.values[0].intersection(roi.geometry.values[0])
        intersection_bounds = intersection_geometry.bounds
        x_len = intersection_bounds[2] - intersection_bounds[0]
        y_len = intersection_bounds[3] - intersection_bounds[1]
        if y_len/x_len > slope_threshold:
            naive_intersect['min_lat'] = intersection_geometry.bounds[1]
            naive_intersect['max_lat'] = intersection_geometry.bounds[3]
            return naive_intersect
        else:
            return None
    else:
        return None

gd_track_subset = j3_s6_intersect(gd_track, buffered_roi)

if gd_track_subset is None:
    raise Exception('No intersection found.')

roi_gdf = gpd.GeoDataFrame(buffered_roi).rename({0: 'geometry'}, axis=1)
roi_gdf = roi_gdf.set_geometry('geometry')
roi_gdf['name'] = RESERVOIR_NAME

plot_lims = buffered_roi.buffer(0.02).bounds
minx = plot_lims.values[0][0]
maxx = plot_lims.values[0][2]
miny = plot_lims.values[0][1]
maxy = plot_lims.values[0][3]

roi.hvplot(color='gray') * gd_track_subset.hvplot(
     by='track', xlim=(minx, maxx), ylim=(miny, maxy), line_width=3, cmap='viridis'
).opts(aspect='equal', xlabel='longitude', ylabel='latitude', title=f"{RESERVOIR_NAME} - Sentinel-6 P|asses")
```

```python
def detect_missing_data(var_FillValue, data):
    if data==var_FillValue:
        data_Flag=1
    else:
        data_Flag=0
    return data_Flag    

def process_s6_file(datapath, lat_boundary):
    datapath = Path(datapath)
    cycno = int(datapath.name.split('_')[-5])
    data=h5py.File(datapath)
    elevations = []

    ## Variable Extraction
    time_01=data['data_01']['time'][:]
    iono_cor_gim=data['data_01']['ku']['iono_cor_gim'][:]
    poletide=data['data_01']['pole_tide'][:]
    solid_earth_tide=data['data_01']['solid_earth_tide'][:]               

    indx_20hzIn01hz = data['data_20']['ku']['index_1hz_measurement'][:]
    time20ku=data['data_20']['ku']['time'][:]

    lon=data['data_20']['ku']['longitude'][:]
    lat=data['data_20']['ku']['latitude'][:]

    mod_dry_tropo_cor_meas=data['data_20']['ku']['model_dry_tropo_cor_measurement_altitude'][:]
    mod_wet_tropo_cor_meas=data['data_20']['ku']['model_wet_tropo_cor_measurement_altitude'][:]
    altimetry=data['data_20']['ku']['altitude'][:]
    range_ocog=data['data_20']['ku']['range_ocog'][:]
    sig0_ocog=data['data_20']['ku']['sig0_ocog'][:]


    ## FillValue Extraction
    iono_cor_gim_FillValue=data['data_01']['ku']['iono_cor_gim'].attrs['_FillValue'][0]
    poletide_FillValue=data['data_01']['pole_tide'].attrs['_FillValue'][0]
    solid_earth_tide_FillValue=data['data_01']['solid_earth_tide'].attrs['_FillValue'][0]               

    indx_20hzIn01hz_FillValue = data['data_20']['ku']['index_1hz_measurement'].attrs['_FillValue'][0]

    lon_FillValue=data['data_20']['ku']['longitude'].attrs['_FillValue'][0]
    lat_FillValue=data['data_20']['ku']['latitude'].attrs['_FillValue'][0]

    mod_dry_tropo_cor_meas_FillValue=data['data_20']['ku']['model_dry_tropo_cor_measurement_altitude'].attrs['_FillValue'][0]
    mod_wet_tropo_cor_meas_FillValue=data['data_20']['ku']['model_wet_tropo_cor_measurement_altitude'].attrs['_FillValue'][0]
    altimetry_FillValue=data['data_20']['ku']['altitude'].attrs['_FillValue'][0]
    range_ocog_FillValue=data['data_20']['ku']['range_ocog'].attrs['_FillValue'][0]
    sig0_ocog_FillValue=data['data_20']['ku']['sig0_ocog'].attrs['_FillValue'][0]
    
    
    ## Scale-factor Extraction
    iono_cor_gim_Scale=data['data_01']['ku']['iono_cor_gim'].attrs['scale_factor'][0]        
    poletide_Scale=data['data_01']['pole_tide'].attrs['scale_factor'][0]
    solid_earth_tide_Scale=data['data_01']['solid_earth_tide'].attrs['scale_factor'][0]               

    lon_Scale=data['data_20']['ku']['longitude'].attrs['scale_factor'][0]
    lat_Scale=data['data_20']['ku']['latitude'].attrs['scale_factor'][0]

    mod_dry_tropo_cor_meas_Scale=data['data_20']['ku']['model_dry_tropo_cor_measurement_altitude'].attrs['scale_factor'][0]
    mod_wet_tropo_cor_meas_Scale=data['data_20']['ku']['model_wet_tropo_cor_measurement_altitude'].attrs['scale_factor'][0]
    altimetry_Scale=data['data_20']['ku']['altitude'].attrs['scale_factor'][0]
    range_ocog_Scale=data['data_20']['ku']['range_ocog'].attrs['scale_factor'][0]
    sig0_ocog_Scale=data['data_20']['ku']['sig0_ocog'].attrs['scale_factor'][0]        
    
    
    ## Add-offset Extraction
    altimetry_Add=data['data_20']['ku']['altitude'].attrs['add_offset'][0]        
    range_ocog_Add=data['data_20']['ku']['range_ocog'].attrs['add_offset'][0]
            
    
    valid_lat_index = np.logical_and(lat*lat_Scale>=lat_boundary[0], lat*lat_Scale<=lat_boundary[1])                       

    mjd_20hz=time20ku*1/86400+51544;
    ## Flags and Correction
    for i in np.arange(len(time20ku)):
        if valid_lat_index[i]==True:
            
            index_20hzIn01hz_Flag=detect_missing_data(indx_20hzIn01hz_FillValue, indx_20hzIn01hz[i])
            if index_20hzIn01hz_Flag==1:
                continue
            
            index_01hz = indx_20hzIn01hz[i]
            pTide_Flag=detect_missing_data(poletide_FillValue,poletide[index_01hz])
            sTide_Flag=detect_missing_data(solid_earth_tide_FillValue,solid_earth_tide[index_01hz])
            iono_Flag=detect_missing_data(iono_cor_gim_FillValue,iono_cor_gim[index_01hz])
                    
            dry_Flag=detect_missing_data(mod_dry_tropo_cor_meas_FillValue,mod_dry_tropo_cor_meas[i])                    
            wet_Flag=detect_missing_data(mod_wet_tropo_cor_meas_FillValue,mod_wet_tropo_cor_meas[i])
                                    
            lat_Flag = detect_missing_data(lat_FillValue,lat[i])                
            alt_Flag = detect_missing_data(altimetry_FillValue,altimetry[i])
            range_Flag = detect_missing_data(range_ocog_FillValue,range_ocog[i])
            #sig0_Flag = detect_missing_data(sig0_ocog_FillValue,sig0_ocog[i])
                                            
            Flags=dry_Flag+wet_Flag+iono_Flag+sTide_Flag+pTide_Flag+alt_Flag+lat_Flag+range_Flag #+sig0_Flag   
                        
            if Flags == 0: # and range_ocean_flag[j] == 0: # and surf_class[j]==2: ##
                pTide_var = poletide[index_01hz]*poletide_Scale
                sTide_var = solid_earth_tide[index_01hz]*solid_earth_tide_Scale
                iono_var = iono_cor_gim[index_01hz]*iono_cor_gim_Scale
                dry_var = mod_dry_tropo_cor_meas[i]*mod_dry_tropo_cor_meas_Scale
                wet_var = mod_wet_tropo_cor_meas[i]*mod_wet_tropo_cor_meas_Scale
                lat_var = lat[i]*lat_Scale
                lon_var = lon[i]*lon_Scale
                alt_var = altimetry[i]*altimetry_Scale+altimetry_Add
                range_var = range_ocog[i]*range_ocog_Scale+range_ocog_Add
                sig0_var = sig0_ocog[i]*sig0_ocog_Scale
                correction=pTide_var+sTide_var+iono_var+dry_var+wet_var                             

                hgt_var=alt_var-(correction+range_var)
                mjd_var=mjd_20hz[i]

                # text_file.write("%4i\t%4i\t%4i\t%4i\t%4i\t%4i\t%4i\t%4i\t%4i\t%4i\t%20.6f\t%20.6f\t%20.6f\t%20.6f\t%10.3f\n" %(dry_Flag, wet_Flag, iono_Flag, sTide_Flag, pTide_Flag, alt_Flag, lat_Flag, range_Flag, Flags, cycno, mjd_var, lon_var, lat_var, hgt_var, sig0_var))
                elevations.append((dry_Flag, wet_Flag, iono_Flag, sTide_Flag, pTide_Flag, alt_Flag, lat_Flag, range_Flag, Flags, cycno, mjd_var, lon_var, lat_var, hgt_var, sig0_var))
    
    extracted_elevations = pd.DataFrame(
        elevations,
        columns=[
            'dry_Flag', 'wet_Flag', 'iono_Flag', 'sTide_Flag', 'pTide_Flag', 'alt_Flag', 'lat_Flag', 'range_Flag', 'Flags', 'cycno', 'mjd', 'lon', 'lat', 'hgt', 'sig0'
        ]
    )
    del data
    return extracted_elevations
```

```python
# ----- Functions of processing -----
# Deoutlier functions
def iqr_deoutlier(cyc_hgt_profile):
    IQR = iqr(cyc_hgt_profile, nan_policy='omit')
    high_prct = np.nanquantile(cyc_hgt_profile, 0.75)
    low_prct = np.nanquantile(cyc_hgt_profile, 0.25)     
    index_retain_iqr = np.logical_and(cyc_hgt_profile>low_prct-1.5*IQR, cyc_hgt_profile<high_prct+1.5*IQR).flatten()   

    return index_retain_iqr

def kmean_water_cluster(cyc_hgt_profile, cyc_sig_profile, hgt_cyc_range_thrd, hgt_cyc_std_thrd):
    
    hgt_cyc_range = np.nanmax(cyc_hgt_profile) - np.nanmin(cyc_hgt_profile)
    while hgt_cyc_range > hgt_cyc_range_thrd:
        kmeans_cluster = KMeans(n_clusters=2, random_state=42, n_init=10).fit_predict(cyc_hgt_profile)                
        cyc_hgt_profile = cyc_hgt_profile[kmeans_cluster==mode(kmeans_cluster, keepdims=True)[0]]
        cyc_sig_profile = cyc_sig_profile[kmeans_cluster==mode(kmeans_cluster, keepdims=True)[0]]
        hgt_cyc_range = np.nanmax(cyc_hgt_profile) - np.nanmin(cyc_hgt_profile)

    hgt_cyc_mean = np.nanmean(cyc_hgt_profile)
    hgt_cyc_std = np.nanstd(cyc_hgt_profile)
    hgt_cyc_errmean = cyc_hgt_profile[:,0] - hgt_cyc_mean
    while hgt_cyc_std > hgt_cyc_std_thrd:
        if np.count_nonzero(~np.isnan(cyc_hgt_profile))==2:
            break
        hgt_cyc_errmean = cyc_hgt_profile - hgt_cyc_mean
        cyc_hgt_profile = cyc_hgt_profile[np.abs(hgt_cyc_errmean)!=np.nanmax(np.abs(hgt_cyc_errmean))]
        cyc_sig_profile = cyc_sig_profile[np.abs(hgt_cyc_errmean)!=np.nanmax(np.abs(hgt_cyc_errmean))]
        hgt_cyc_std = np.nanstd(cyc_hgt_profile)
        hgt_cyc_mean = np.nanmean(cyc_hgt_profile)
    
    return cyc_hgt_profile, cyc_sig_profile

def get_time_series(input_data, index_retain, hgt_cyc_range_thrd, hgt_cyc_std_thrd):
    input_data = input_data[index_retain,:]
    cycno_list = input_data[:,0]
    uniq_cycno = np.unique(cycno_list)

    ct_cyc=0
    FinalSeries=np.empty((len(uniq_cycno),8))
    FinalSeries[:]=np.nan
    for cycno in uniq_cycno[:]:
        
        index_cyc = cycno_list==cycno
        cyc_data = input_data[index_cyc,:]

        mjd = cyc_data[:,1].reshape(-1,1)
        lon = cyc_data[:,2].reshape(-1,1)
        lat = cyc_data[:,3].reshape(-1,1)
        hgt = cyc_data[:,4].reshape(-1,1)
        sig = cyc_data[:,5].reshape(-1,1)    
               
        hgt, sig = kmean_water_cluster(hgt, sig, hgt_cyc_range_thrd, hgt_cyc_std_thrd)

        cyc_time = (np.nanmean(mjd)+2108-50000)/365.25 +1990
        cyc_lon = np.nanmean(lon)
        cyc_lat = np.nanmean(lat)
        
        cyc_hgt = np.nanmean(hgt)
        #cyc_unc_hgt = uncertainty(hgt)  
        cyc_std_hgt = np.nanstd(hgt)      
        cyc_sig = np.nanmean(sig)
        cyc_retain_rate = hgt.shape[0] / cyc_data.shape[0]
           
        FinalSeries[ct_cyc,:] = [cycno, cyc_time, cyc_lon, cyc_lat, cyc_hgt, cyc_std_hgt, cyc_sig, cyc_retain_rate]       
            
        ct_cyc=ct_cyc+1    
    
    return FinalSeries

def get_time_series_egm08(ip, FinalSeries):
    ## Time-series of water levels (convert to w.r.t. EGM-2008)
    N=ip((FinalSeries[:,3],FinalSeries[:,2]))
    FinalSeries[:,4]=FinalSeries[:,4]-N
    
    egm08_timeseries = FinalSeries[:,[0,4,5,1,2,3,7]]

    return egm08_timeseries
```

```python
def okeowo_2017_s6(
        extracted_elevations_fp, min_lat, max_lat, 
        hgt_cyc_range_thrd=5, hgt_cyc_std_thrd=0.3, 
        egm2008_fp="/tiger1/pdas47/tmsosPP/data/geoid/geoidegm2008grid.mat",
        cycle_retention_z_score_threshold = 2
    ):
    lat_boundary=[min_lat, max_lat]

    # Data Import
    # 1. EGM2008
    lonbp = io.loadmat(egm2008_fp)['lonbp']
    latbp = io.loadmat(egm2008_fp)['latbp']
    grid=io.loadmat(egm2008_fp)['grid']

    ip=RegularGridInterpolator(points=(latbp.flatten(),lonbp.flatten()), values=grid, bounds_error=False, fill_value=np.nan)

    # 4. Altimetry data
    input_data = np.loadtxt(extracted_elevations_fp)
    if len(input_data)==0:
        print(f"No data in {extracted_elevations_fp}")
        return None
    
    input_data = input_data[:,9:]
    lat_all=input_data[:,3]
    index_lat_range = np.logical_and(lat_all>lat_boundary[0], lat_all<lat_boundary[1])
    input_data=input_data[index_lat_range,:]

    lon_all = input_data[:,2].reshape(-1,1)
    lat_all = input_data[:,3].reshape(-1,1)
    hgt_all = input_data[:,4].reshape(-1,1)
    index_retain_iqr = iqr_deoutlier(hgt_all)

    FinalSeries_iqr = get_time_series(input_data, index_retain_iqr, hgt_cyc_range_thrd, hgt_cyc_std_thrd)

    egm08_timeseries = get_time_series_egm08(ip, FinalSeries_iqr)
    
    df = pd.DataFrame(
        egm08_timeseries,
        columns=['cycno', 'hgt', 'std_hgt', 'time', 'lon', 'lat', 'cycle_retain_rate']
    )

    df['z_score_cycle_retension'] = np.abs((df['cycle_retain_rate'] - df['cycle_retain_rate'].mean()) / df['cycle_retain_rate'].std())
    df = df[df['z_score_cycle_retension'] <= cycle_retention_z_score_threshold]

    return df
```

```python
def s6_altimeter(
        roi, start_date, end_date,
        intersection_length_threshold=300, # meters. jason class altimeters require at least 300 m of intersection with the ROI.
        s6_tracks_fp = Path("/tiger1/pdas47/tmsosPP/data/j3_s6_tracks/j3_s6_tracks.geojson"),
        s6_raw_dir = Path("/tiger1/pdas47/tmsosPP/data/s6/raw"),
        s6_extracted_dir = Path("/tiger1/pdas47/tmsosPP/data/s6/extracted"),
        dask_client = None,
        gd_track_slope_threshold = 0.01,
        hgt_cyc_range_thrd = 5,
        hgt_cyc_std_thrd = 0.3,
        cycle_retention_z_score_threshold = 2
    ):
    tmsos_id = roi['tmsos_id'].values[0]
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)

    # create output directory if it doesn't exist
    s6_extracted_dir.mkdir(exist_ok=True)

    # get lat-lon boundary
    gd_track = gpd.read_file(s6_tracks_fp)
    gd_track_subset = j3_s6_intersect(gd_track, roi, slope_threshold=gd_track_slope_threshold)

    # filter by intersection length threshold
    nadir_intersects = gpd.GeoDataFrame(geometry=gd_track_subset.intersection(roi.geometry.values[0]).explode(index_parts=False)) # explode = split multi-geometry into single geometries
    nadir_intersects['intersect_length_m'] = nadir_intersects.to_crs(roi.estimate_utm_crs()).length
    nadir_intersects = nadir_intersects[nadir_intersects['intersect_length_m'] > intersection_length_threshold]
    nadir_intersects['pass_num'] = gd_track_subset['track'].loc[nadir_intersects.index]
    nadir_intersects['min_lat'] = nadir_intersects.bounds['miny']
    nadir_intersects['max_lat'] = nadir_intersects.bounds['maxy']
    
    s6_timeseries_all = []
    # for each pass in ROI, extract data
    for idx, row in nadir_intersects.iterrows():
        extracted_elevations_fp = s6_extracted_dir / f"S6A_{tmsos_id}_{row['pass_num']}_{row['min_lat']}_{row['max_lat']}_{start_date:%Y%m%d}_{end_date:%Y%m%d}_info.txt"
        pass_num = int(row['pass_num'])
        lat_min = row['min_lat']
        lat_max = row['max_lat']
        intersect_length = row['intersect_length_m']
        print(extracted_elevations_fp)

        if extracted_elevations_fp.exists() and extracted_elevations_fp.stat().st_size > 0:  # skip if the processed file already exists and has data
            print(f"Skipping {extracted_elevations_fp} as it already exists")
        else:
            # obtain file paths that match the pass number
            s6_nadir_fps = list(Path(s6_raw_dir).glob(f'S6A_*_{pass_num:03}_*.nc'))

            # filter by start and end date
            s6_nadir_fps = filter(
                lambda fp: pd.to_datetime(fp.name.split('_')[-3], format='%Y%m%dT%H%M%S') >= pd.to_datetime(start_date) \
                    and pd.to_datetime(fp.name.split('_')[-3], format='%Y%m%dT%H%M%S') <= pd.to_datetime(end_date),
                s6_nadir_fps
            )
            
            # process the files. If dask client is provided, process the files parallely
            futures = []
            if dask_client is None:
                with open(extracted_elevations_fp, "w") as text_file:
                    print(f"writing to {extracted_elevations_fp}")
                    extracted_elevations = []
                    for fp in s6_nadir_fps:
                        processed_elevation_info = process_s6_file(fp, lat_boundary=[lat_min, lat_max])
                        extracted_elevations.append(processed_elevation_info)
                    s6_extracted_elevations = pd.concat(extracted_elevations)
                    s6_extracted_elevations.to_csv(text_file, index=False, sep='\t', header=False)
            else:
                # dask client is provided. Process the files parallely
                futures = []
                for fp in s6_nadir_fps:
                    future = dask_client.submit(process_s6_file, fp, lat_boundary=[lat_min, lat_max])
                    futures.append(future)
                extracted_elevations = dask_client.gather(futures)
                # write the processed info into text file
                with open(extracted_elevations_fp, "w") as text_file:
                    s6_extracted_elevations = pd.concat(extracted_elevations)
                    s6_extracted_elevations.to_csv(text_file, index=False, sep='\t', header=False)

        # process the extracted elevations to get final time series
        s6_timeseries = okeowo_2017_s6(
            extracted_elevations_fp, lat_min, lat_max, 
            hgt_cyc_range_thrd, hgt_cyc_std_thrd, 
            egm2008_fp="/tiger1/pdas47/tmsosPP/data/geoid/geoidegm2008grid.mat",
            cycle_retention_z_score_threshold = cycle_retention_z_score_threshold
        )
        if s6_timeseries is None:
            print(f"No data in {extracted_elevations_fp}")
            continue
        s6_timeseries['lon'] = s6_timeseries['lon'].apply(lambda lon: lon - 360 if lon > 180 else lon)
        s6_timeseries['pass_num'] = pass_num
        s6_timeseries['intersect_length_m'] = intersect_length
        s6_timeseries_all.append(s6_timeseries)

    return pd.concat(s6_timeseries_all)
```
